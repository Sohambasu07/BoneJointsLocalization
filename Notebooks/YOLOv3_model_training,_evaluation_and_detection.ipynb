# Detection code

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from PIL import Image
# Load Pre-trained Yolo Algorithm
yolo = cv2.dnn.readNet("/content/drive/MyDrive/yolo_custom_model_Training/backup/yolov3_custom_final.weights", "/content/drive/MyDrive/yolo_custom_model_Training/darknet/cfg/yolov3_custom.cfg")
# Read the Coco Dataset
classes = []
with open("/content/drive/MyDrive/yolo_custom_model_Training/custom_data/classes.names", 'r') as f:
    classes = f.read().splitlines()
print("Number of Class names", len(classes))
count = 0
for k in range(1,2):
    # Read Image, Dimension of Image, Resize it
    DirPathObj = "/content"
    FilesObj = f"pic4.jpg"
    #/content/drive/MyDrive/yolo_custom_model_Training/custom_k/img_(k).jpg
    #/content/drive/MyDrive/custom_test/image  (1).png
    imgPathObj = os.path.join(DirPathObj, FilesObj)
    img = np.array(Image.open(imgPathObj))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, None, fx=1, fy=1)
    height, width, channels = img.shape
    img_copy = img.copy()
    # Create Blob
    blob = cv2.dnn.blobFromImage(img, 1/255.0, (416,416), (0, 0, 0), swapRB=False, crop = False)
    # Pass I/O for Yolo
    yolo.setInput(blob)
    output_layers_name = yolo.getUnconnectedOutLayersNames()
    layeroutput = yolo.forward(output_layers_name)
    # Start Detecting
    boxes = []
    confidences = []
    class_ids = []
    for output in layeroutput:
        for detection in output:
            score = detection[5:]
            class_id = np.argmax(score)
            confidence = score[class_id]
            if confidence > 0.25:
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)
    # Remove Extra bounding box
    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.5)
    font = cv2.FONT_HERSHEY_PLAIN
    colors = np.random.uniform(0, 255, size=(len(classes), 3))
    # Flatten Parameters
    if len(indexes) > 0:
        for i in indexes.flatten():
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            confi = str(round(confidences[i], 2))
            x2 = x + w
            y2 = y + h
            cv2.rectangle(img, (x, y), (x2, y2),(255,255,0), 1)
            cv2.putText(img, label + " " + confi, (x, y + 20), font, 2, (255, 255, 255), 2)
            print(confidences[i])
            plt.figure(k)
            plt.imshow(img)
            plt.show()
            cv2.imwrite(f"/content/drive/MyDrive/yolo_custom_model_Training/test_results/pred_Image {k}.png", img)
    else:
      print("Cannot detect joints in ",FilesObj)
      count+=1
```
